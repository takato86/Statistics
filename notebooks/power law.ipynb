```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 100)
```

```python
C = 1.0
ALPHA = 2.0
y_pow = C * x **-ALPHA
y_exp = C * np.exp(-x)
```

指数分布
$$
p(x) = Ce^{-x}
$$
べき乗分布
$$
p(x) = Cx^{-\alpha}
$$

```python
fig = plt.figure()
plt.plot(x, y_pow)
plt.plot(x, y_exp)
plt.legend(["pow","exp"])
```

べき乗分布の方が減少が緩やか

# べき乗分布の両軸を対数化

```python
plt.plot(np.log(x), np.log(y_pow))
plt.title("power distribution after logarithmic conversion")
```

```python
plt.plot(np.log(x), np.log(y_exp))
plt.title("exponential distribution after logarithmic conversion")
```

べき乗分布に従っていた場合，両軸対数のグラフだと右肩下がりの直線となる．
指数分布だとならない．

```python
a = 1
b = 1
y = (a/b) / (x / b)**(a+1) 
```

```python
plt.plot(x, y)
plt.title("Pareto distribution")
```
